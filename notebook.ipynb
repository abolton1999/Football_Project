```python
# load the JSON file as a dictionary
'''
This file contains general game info - important info:

1. Date (year, month, day)
2. Pitch Dimensions?
3. Players (name, number, position)
4. Score

'''
with open(r"Data\g2312135_SecondSpectrum_meta.json", 'r') as f:
    data = json.load(f)
    
# normalize the nested JSON data
normalized_data = pd.json_normalize(data)

# convert the normalized data to a pandas dataframe
meta_data = pd.DataFrame(normalized_data)
```

```python

def player_ids_dict():

    player_ids = {}

    for index in range(len(meta_data['homePlayers'][0])):

        player_ids[meta_data['homePlayers'][0][index]['ssiId']]=[]

    return player_ids
    
```
